```python
import numpy as np  #      
import matplotlib.pyplot as plt  #   
#   x  0  10  100 
x = np.linspace(0, 10, 100)

#  sin(x)  cos(x)
y = np.sin(x)
z = np.cos(x)

#  
plt.figure(figsize=(10, 5))  #  
plt.plot(x, y, label='sin(x)')  #  sin(x)
plt.plot(x, z, label='cos(x)')  #  cos(x)
plt.title(" sin(x)  cos(x)")  # 
plt.xlabel("x")  #   X
plt.ylabel("y")  #   Y
plt.legend()  #  
plt.grid()  #  
plt.show()  #  
```

```python
import numpy as np  #      
import matplotlib.pyplot as plt  #   

#   55     1  10
matrix = np.random.randint(1, 11, size=(5, 5))

#  
print(" 55:")
print(matrix)

#   
print("\n :", np.mean(matrix))

#    
print(" :", np.max(matrix))
print(" :", np.min(matrix))

#    (axis=0  , axis=1  )
print("  :", np.sum(matrix, axis=0))

plt.figure(figsize=(8, 6))
plt.imshow(matrix, cmap='viridis')  #   
plt.colorbar()  #   
plt.title("  ")

#    
for i in range(5):  #  
    for j in range(5):  #  
        plt.text(j, i, matrix[i, j], ha='center', va='center', color='white')

plt.show()
```

```python
import numpy as np  #      
import matplotlib.pyplot as plt  #   

#    3   
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 1.  y = x
x = np.linspace(0, 10, 50)
axes[0].plot(x, x**2)
axes[0].set_title("y = x")  # 
axes[0].grid()

# 2.    
x_points = np.random.rand(50)  # 50    0  1
y_points = np.random.rand(50)
axes[1].scatter(x_points, y_points)
axes[1].set_title(" ")
axes[1].grid()

# 3.  
categories = ['A', 'B', 'C']
values = [3, 7, 2]
axes[2].bar(categories, values)
axes[2].set_title(" A, B, C")
axes[2].grid()

plt.tight_layout()  #   
plt.show()
```
